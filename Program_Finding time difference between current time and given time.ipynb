```r
# Installing the necessary packages:

install.packages("tidyverse")
library(tidyverse)
library(lubridate)
```

```r
# Creating two columns for given_time and current_time respectively:

given_time <- c("11:00:00", "07:10:12", "16:08:04")
current_datetime <- c(now())                        #now() functions gives the current time (in UTC time zone by default)

df <- data.frame(given_time, current_datetime)
```

```r
# Previewing the dataframe:

head(df)
```

```r
# Seperating datetime column in date and time:

df <- df %>%
    separate(current_datetime, c("current_date", "current_time"), sep=" ")
```

```r
head(df)
```

```r
df <- mutate(df, given_time = as.character(given_time))
```

```r
head(df)
```

```r
df <- df %>% mutate(given_time = hms::as.hms(given_time), current_time= hms::as.hms(current_time))
```

```r
head(df)
```

```r
df$time_diff <- difftime(df$given_time,df$current_time)
```

```r
head(df)
```

```r
# The above column of time_diff represents the time difference between the two columns.
```
